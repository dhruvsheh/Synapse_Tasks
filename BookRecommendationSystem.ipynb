```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
books = pd.read_csv('BX-Books.csv',sep=";",on_bad_lines='skip', encoding='latin-1',quoting=3)
```

```python
books=books[['"ISBN"', '"Book-Title"', '"Book-Author"', '"Year-Of-Publication"',
       '"Publisher"','"Image-URL-L"']]
books.head(5)
```

```python
books.rename(columns={'"ISBN"':'isbn','"Book-Title"':'title','"Book-Author"':'Author','"Year-Of-Publication"':'Year','"Publisher"':'Publisher','"Image-URL-L"':'image_url'}, inplace= True)
```

```python
books.head()
```

```python
users=pd.read_csv('BX-Users.csv',encoding='latin-1',sep=";",on_bad_lines='skip',quoting=3)
```

```python
users.rename(columns={'"User-ID"':'ID','"Location"':'location','"Age"':'Age'},inplace=True)
```

```python
ratings=pd.read_csv('BX-Book-Ratings.csv',encoding='latin-1',sep=";",on_bad_lines='skip',quoting=3)
ratings.info()
```

```python
print(books.shape)
print(users.shape)
print(ratings.shape)
```

```python
ratings.head()
```

```python
ratings.rename(columns={'"User-ID"':'user_id','"ISBN"':'isbn','"Book-Rating"':'rating'},inplace=True)
```

```python
ratings.head()
```

```python
ratings['user_id'].value_counts()
```

```python
x=ratings['user_id'].value_counts()>200
#x[x] filters the series x to include only the elements where the value is True.
x[x].shape
```

```python
y=x[x].index
y
```

```python
ratings=ratings[ratings['user_id'].isin(y)]
ratings.head()
```

```python
ratings_and_books=ratings.merge(books,on='isbn')
ratings_and_books.head()
```

```python
num_of_ratings=ratings_and_books.groupby('title')['rating'].count().reset_index()
num_of_ratings.rename(columns={'rating':'num_of_ratings'},inplace=True)
num_of_ratings.head()
```

```python
final_rating_1=ratings_and_books.merge(num_of_ratings,on='title')
final_rating_1.head()
```

```python
final_rating_1.shape
```

```python
final_rating_1 = final_rating_1.replace('"', '', regex=True)
final_rating_1
```

```python
final_rating_1=final_rating_1[final_rating_1['num_of_ratings']>=50]
```

```python
final_rating_1.shape
```

```python
final_rating_1.head()
```

```python
final_rating_1.sample(10)
```

```python
final_rating_1.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating_1.shape
```

```python
final_rating_1
```

```python
pivot_book=final_rating_1.pivot_table(columns='user_id',index='title',values='rating')
pivot_book
```

```python
pivot_book.fillna(0,inplace=True)
pivot_book
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(pivot_book)
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute') #brute calculates the squared distances from each query feature vector to each reference feature vector in the training data
model.fit(book_sparse)
```

```python
distance, suggestion=model.kneighbors(pivot_book.iloc[23,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
  print(pivot_book.index[suggestion[i]])
```

```python
book_name=pivot_book.index
```

```python
import pickle
import os

# Create the 'meow' directory if it doesn't exist
if not os.path.exists('meow'):
    os.makedirs('meow')
pickle.dump(model,open('meow/model.pkl','wb')) #wb-write binary mode as we want to write it
pickle.dump(book_name,open('meow/book_name.pkl','wb'))
pickle.dump(final_rating_1,open('meow/final_rating_1.pkl','wb'))
pickle.dump(pivot_book,open('meow/pivot_book.pkl','wb'))
#The pickle module can transform a complex object into a byte stream and it can transform the byte stream into an object with the same internal structure.
#Perhaps the most obvious thing to do with these byte streams is to write them onto a file, but it is also conceivable to send them across a network or store them in a database.
#The shelve module provides a simple interface to pickle and unpickle objects on DBM-style database files
```

```python
def recommend_book(book_name):
    books_list = []
    book_id = np.where(pivot_book.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(pivot_book.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6 )
    for i in range(len(suggestion)):
            books = pivot_book.index[suggestion[i]]
            for j in books:
                print(j)
```
